```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Dee-M123/SQL-Mini-Project/refs/heads/main/DATA/RAW/Coffee%20Shop%20Sales%20Analysis_dataset.csv")
```

```python
df.head(3)
```

```python
df.columns = (df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_'))
```

```python
df.head(2)
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

**We have established that we have duplicated rows and no null values, so we can create our tables for SQL**


```python
stores = df[['store_id', 'store_location']].drop_duplicates()
```

```python
stores.shape
```

```python
stores.to_csv("stores_sql.csv", index=False)
```

```python
products = df[['product_id', 'product_category', 'product_type',
     'product_detail', 'size', 'unit_price']
].drop_duplicates()
```

```python
products.to_csv("products_sql.csv", index=False)
```

```python
transactions = df[['transaction_id', 'transaction_date', 'transaction_time',
     'store_id', 'product_id',"transaction_qty", 'total_bill',
     'hour', 'day_name', 'month', 'month_name', 'day_of_week']
]
```

```python
transactions.to_csv("transactions_sql.csv", index=False)
```
