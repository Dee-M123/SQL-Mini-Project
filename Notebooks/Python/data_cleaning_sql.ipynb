```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/Dee-M123/SQL-Mini-Project/refs/heads/main/DATA/RAW/Coffee%20Shop%20Sales%20Analysis_dataset.csv")
```

```python
df.head(3)
```

```python
df.columns = (df.columns
      .str.strip()
      .str.lower()
      .str.replace(' ', '_'))
```

```python
df.columns
```

**Missing values and duplicates**

**Numeric columns and dates**

```python
# ID columns
id_cols = ["transaction_id", "store_id", "product_id"]
for col in id_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce").astype("Int64")

```

```python
# Date
df["transaction_date"] = pd.to_datetime(
    df["transaction_date"], errors="coerce"
)

df['transaction_date'] = pd.to_datetime(
    df['transaction_date'], dayfirst=True
).dt.strftime('%Y-%m-%d')
```

```python
# Numeric business fields
df["transaction_qty"] = pd.to_numeric(df["transaction_qty"], errors="coerce")
df["unit_price"] = pd.to_numeric(df["unit_price"], errors="coerce")
df["total_bill"] = pd.to_numeric(df["total_bill"], errors="coerce")
```

```python
# Time
df["transaction_time"] = pd.to_datetime(
    df["transaction_time"], format="%H:%M:%S", errors="coerce"
).dt.time
```

```python
df.to_csv("coffee_sales_sql_ready.csv", index=False)
```
