```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**H1:** Certain store locations generate  higher total revenue than others.

```python
df = pd.read_csv("https://raw.githubusercontent.com/Dee-M123/SQL-Mini-Project/refs/heads/main/DATA/SQL-Python-Visualisation-CSV/store_revenue.csv")

plt.figure()
plt.bar(df["store_location"], df["total_revenue"])
plt.xlabel("Store Location")
plt.ylabel("Total Revenue")
plt.title("Hypothesis 1: Revenue by Store Location")
plt.show()
```

**H2:** Coffee products contribute more to overall revenue than non-coffee products

```python
df = pd.read_csv("https://raw.githubusercontent.com/Dee-M123/SQL-Mini-Project/refs/heads/main/DATA/SQL-Python-Visualisation-CSV/product_group_revenue.csv")

plt.figure()
plt.bar(df["product_group"], df["total_revenue"])
plt.xlabel("Product Group")
plt.ylabel("Total Revenue")
plt.title("Hypothesis 2: Revenue by Product Group")
plt.show()
```

**H3:** Sales revenue peaks during specific hours of the day.

```python
df = pd.read_csv("https://raw.githubusercontent.com/Dee-M123/SQL-Mini-Project/refs/heads/main/DATA/SQL-Python-Visualisation-CSV/hourly_revenue.csv")

df = df.sort_values("hour")

plt.figure(figsize=(12, 6))

plt.bar(df["hour"],df["total_revenue"],edgecolor="black")
plt.xlabel("Hour of Day", fontsize=12)
plt.ylabel("Total Revenue ($)", fontsize=12)
plt.title("Hypothesis 3: Total Revenue by Hour of Day", fontsize=14, fontweight="bold")
plt.xticks(df["hour"])  # show every hour
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
